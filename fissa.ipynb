```python
import decimal
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
import warnings
```

```python
def newtonrhapson(func, derivitive, x0, max_iter = 300):
    if derivitive(x0) == 0:
        warnings.warn('derivitive is zero')
    outputs_list = []
    xk = x0
    for i in range(max_iter):
        xk = xk - func(xk) / derivitive(xk)
        outputs_list.append(xk)
        
    return outputs_list

def refined_newton(func, derivitive, second_derivitive, x0, M, max_iter = 300):
    outputs_list = []
    xk = x0
    for i in range(max_iter):
        xk = xk - (func(xk) * derivitive(xk)) / (derivitive(xk)**2 -(1/M)*func(xk)*second_derivitive(xk))
        outputs_list.append(xk)
    
    return outputs_list

def linearisation(func, derivitive, x):
    a = derivitive(x)
    b = func(x) - a*x
    def lijn(x):
        return a*x + b
    
    return lijn
```

```python
f = lambda x: x**3. - 5.
fprime = lambda x: 3.*x**2.
fsecond = lambda x: 6.*x

F = lambda x: f(x)/ np.sqrt(fprime(x))
Fprime = lambda x: (fprime(x)**2 - (1./3.)*f(x)*fsecond(x))/(fprime(x)**(3./2.))

x_range = np.linspace(0, 10, 200)
xax = np.zeros(len(x_range))

x0 = 5
points = newtonrhapson(f, fprime, x0, max_iter = 10)

special = linearisation(f, fprime, x0)
kurwa = special(x_range)
lijnnen = []
for p in points:
    temp = linearisation(f, fprime, p)
    lijnnen.append(temp(x_range))
    
lijnnen = np.array(lijnnen)
```
